```python
from itertools import permutations, repeat, chain
import matplotlib.pyplot as plt
```

```python
# The 5 teams from the Atlantic conference.
T = list(range(5))
```

```python
# Generate all possible pairs of teams and repeat it once.
G = list(chain.from_iterable(repeat(list(permutations(T, 2)), 2)))
len(G)
```

```python
# Generate a sequence of rounds that represent the number of times all teams
# would have to travel in order for every team to complete all of their matches.
def sample_R(G_reduce):
    R = []
    G_reduce = list(G_reduce)
    while len(G_reduce) > 0:
        R_i = []
        teams_playing = set()
        for g in G_reduce:
            if len(set(g).intersection(teams_playing)) == 0:
                R_i.append(g)
                teams_playing.update(set(g))
        R.append(tuple(R_i))
        for g in R_i:
            G_reduce.remove(g)
    return R
```

```python
unique_R = set()
r_len = []
i = 0

for Gp in permutations(G):
    i += 1
    unique_R.add(tuple(sample_R(list(Gp))))
    r_len.append(len(unique_R))

    if i > 1_000_000:
        break
```

```python
plt.plot(r_len)
plt.title("Number of unique round sequences against permutations of Game set")
plt.ylabel("Unique R")
plt.xlabel("Permutations of G")
```
